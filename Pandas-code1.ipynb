```python
unique_pageids=list(from_csv.Page_total_likes.unique())

unique_upload_types=list(from_csv.upload_type.unique())
print(unique_upload_types)

#print unique_pageids
print ("Number of uniwue pages: ",len(unique_pageids)) 
print("Number of rows in from_csv: ",len(from_csv.index))
#print(from_csv.columns)
for i in xrange(1,len(from_csv.index)+1):
    from_csv.loc[i,'page_id']=unique_pageids.index(from_csv.loc[i,'Page_total_likes'])+1
    
#print(from_csv.head())
#print(from_csv.tail())
from_csv.to_csv('C:\\Users\\bodap\\OneDrive\\Documents\\Papers_WinterBreak\\Pandas\\cleaned_facebook.csv', 
                sep=',', encoding='utf-8')
```

```python
photos=np.zeros((1,len(unique_pageids)));
status=np.zeros((1,len(unique_pageids)));
links=np.zeros((1,len(unique_pageids)));
videos=np.zeros((1,len(unique_pageids)));
total_posts=np.zeros((1,len(unique_pageids)));

page_ids=range(1,len(unique_pageids)+1)
for ind,page_id in enumerate(unique_pageids):
    for i in xrange(1,len(from_csv.index)+1):
        if(from_csv.loc[i,'Page_total_likes']== page_id):
            if(from_csv.loc[i,'upload_type'] =='Photo'):
                photos[0,ind]=photos[0,ind]+1;
            elif(from_csv.loc[i,'upload_type'] =='Status'):
                status[0,ind]=status[0,ind]+1;
            elif(from_csv.loc[i,'upload_type'] =='Link'):
                links[0,ind]=links[0,ind]+1;
            else:
                videos[0,ind]=videos[0,ind]+1
for ind in xrange(0,len(unique_pageids)):
    total_posts[0,ind]=photos[0,ind]+status[0,ind]+links[0,ind]+videos[0,ind];
```

```python
d={'Page_total_likes':unique_pageids,
  'photos':photos[0],
   'status':status[0],
   'links':links[0],
   'videos':videos[0],
    'total_posts':total_posts[0]};                

cols=['Page_total_likes','photos','status','links','videos','total_posts'];
new_df=pd.DataFrame(d)
new_df=new_df[cols]
new_df.to_csv('C:\\Users\\bodap\\OneDrive\\Documents\\Papers_WinterBreak\\Pandas\\upload_stats_facebook.csv', 
                sep=',', encoding='utf-8')
```
